```python
#词云图
# 摘要： 当我们手中有一篇文档，比如书籍、小说、电影剧本，若想快速了解其主要内容是什么，则可以采用绘制 WordCloud 词云图，显示主要的关键词（高频词）这种方式，非常方便。
# 本文将介绍常见的英文和中文文本的词云图绘制，以及 Frequency 频词频词云图。
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud ,STOPWORDS
import numpy as np
from PIL import Image
import os
from os import path
```

```python
#获取当前文件路径
d=(path.dir_name(_file_) if "_file_" in locals() else os.getcwd())+"\Data"
#获取文本
text=open(path.join(d,'wordcloud.txt')).read()
#生成词云
wc=WordCloud(scale=2,
             background_color='black',
             max_font_size = 100)
wc.generate_from_text(text)
#显示图像
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.tight_layout()
#存储图像
wc.to_file('./Data/1900_basic.png')
plt.savefig('./Data/1900_basic.png',dpi=200)
plt.show()
```
